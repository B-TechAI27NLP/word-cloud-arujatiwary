```python
# Stopword Removal
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

word_filtered=[]
text_tokenize = word_tokenize(text.lower())

for word in text_tokenize:
    if word not in stop_words:
        word_filtered.append(word)

word_filtered
```

```python
# Punctuation Removal
punc = [",",".","-","'"]

for word in word_tokens:
  for i in punc:
    if word == i:
      word_tokens.remove(word)

word_tokens
```

```python
# Stemming
from nltk.stem import PorterStemmer

words= ["wait", "waiting", "waited", "waits", "listening", "history", "ability", "arrival", "finally", "congratulations",
       "exaggeration", "understandable", "probability", "player","toys", "consumer", "fairly"]

porter = PorterStemmer()
for word in words:
    root = porter.stem(word)
    print(word," - ",root)
```

```python
# Lemmatizing
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
lemmatizer= WordNetLemmatizer()
for word in words:
  lem = lemmatizer.lemmatize(word)
  print(word," - ",lem)
```

```python
from nltk.corpus import stopwords
from gensim.utils import simple_preprocess
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.stem import PorterStemmer
```

```python
sports_keywords = ["game", "match", "score", "team", "win", "loss", "player", "coach",
      "season", "league", "tournament", "championship", "final", "referee",
      "goal", "penalty", "performance", "stats", "lineup", "injury",
      "ranking", "training", "strategy", "fitness", "record", "result",
      "highlight", "schedule", "victory", "draw", "competition", "opponent",
      "substitute", "defense", "offense", "play", "athlete", "event",
      "qualifier", "standings", "venue", "halftime", "whistle"]

finance_keywords = ["market", "stock", "investment", "portfolio", "economy", "inflation",
      "revenue", "profit", "loss", "income", "dividend", "capital",
      "equity", "asset", "liability", "valuation", "bond", "interest",
      "rate", "index", "growth", "return", "fund", "exchange", "trading",
      "shares", "earnings", "forecast", "debt", "liquidity", "margin",
      "risk", "cash", "balance", "budget", "fiscal", "tax", "credit",
      "currency", "regulation", "yield", "benchmark", "derivative",
      "leverage", "inflation", "expense", "volatility", "diversification"]

fashion_keywords = ["style", "trend", "collection", "designer", "runway", "model",
      "couture", "look", "outfit", "fabric", "texture", "pattern",
      "color", "season", "wardrobe", "accessory", "footwear", "apparel",
      "garment", "silhouette", "aesthetic", "vogue", "ensemble",
      "tailoring", "fit", "fashion", "catwalk", "branding", "label",
      "luxury", "streetwear", "casual", "formal", "haute", "vintage",
      "icon", "makeup", "hairstyle", "jewelry", "statement", "collection",
      "editorial", "lookbook", "palette", "minimalist", "bold", "chic",
      "trendsetter", "influencer", "design"]

porter = PorterStemmer()
for key in sports_keywords:
    root = porter.stem(key)
    sports_keywords.remove(key)
    sports_keywords.append(root)

porter = PorterStemmer()
for key in finance_keywords:
    root = porter.stem(key)
    finance_keywords.remove(key)
    finance_keywords.append(root)

porter = PorterStemmer()
for key in fashion_keywords:
    root = porter.stem(key)
    fashion_keywords.remove(key)
    fashion_keywords.append(root)
```

```python
text = input("Enter text: ")
```

```python
def preprocessingtext(text):
  # tokenization and punctuation removal
  text = text.lower()
  words = simple_preprocess(text)

  # stemming
  porter = PorterStemmer()
  for word in words:
      root = porter.stem(word)
      words.remove(word)
      words.append(root)

    # stopword removal
  stop_words = set(stopwords.words("english"))

  word_filtered=[]

  for word in words:
      if word not in stop_words:
          word_filtered.append(word)

  count = 0
  noted = []
  text_dict = {}

  for word in word_filtered:
    if word not in noted:
      noted.append(word)
      for Match in word_filtered:
        if word == Match:
          count += 1
          text_dict[Match] = count
    count = 0

  return text_dict
```

```python
text_dict = preprocessingtext(text)
```

```python
def doc_classification(text_dict):
  count = 0
  for key in text_dict.keys():
    if key in sports_keywords and text_dict[key] >= 1:
      count += 1
      if count > 7:
        count = 0
        print("Sports")

    elif key in finance_keywords and text_dict[key] >= 1:
      count += 1
      if count > 7:
        count = 0
        print("Finance")

    elif key in fashion_keywords and text_dict[key] >= 1:
      count += 1
      if count > 7:
        count = 0
        print("Fashion")
```

```python
doc_classification(text_dict)
```
